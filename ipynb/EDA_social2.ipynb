```python
data3['date'] = pd.to_datetime(data3['date'].astype('str') , format = '%Y-%m-%d')
data3 = data3.set_index(data3['date'])
data3.head()
```

```python
columns = data3.columns
columns = columns[-6:]
columns
```

```python
data3['country_region'].value_counts()
```

```python
countries = ['United States', 'United Kingdom', 'Brazil', 'China', 'India', 'Italy']
#considering important countries
```

```python
for k in columns:

    fig=plt.figure(figsize= (20,10))
    ax=fig.add_subplot(111)
    for i in countries:
        c = data3[data3['country_region'] == i]
        ax.plot(c.index,[0]*len(c), color = 'k')
        sns.lineplot(x = c.index, y = k, data = c,label = i)
    plt.title('MOBILITY: ' + k)
    plt.grid()
    plt.show()
    
```

```python
# All countries experience a negative mobility change from baseline for visiting public places outdoors. Whereas all countries experience a high positive change in mobility from baseline for residential spaces indoors. This concludes that social distancing and lockdown measures were followed. 
```
